### 2.1. Two-view triangulation

````python
import pickle
import os
from pathlib import Path
import torch
import numpy as np

from utils import load_cam_data


def two_view_triangulation(camera_matrix_1, camera_matrix_2, points_1, points_2, score_1, score_2):
    """
        input
        camera_matrix_n: 3 X 4, camera matrix of each camera
        points_n: N x 2, 2D coordinate of N points

        output
        output_dict: dictionary of triangulated points
        key: point_idx from 0 to 18, value: triangulated point, np.array with shape 1 X 3
        output_dict = {
            0: [[x1, y1, z1]],
            1: [[x2, y2, z2]],
            ...
            18: [x18, y18, z18]
          }
    """
    output_dict = {}
    #################### TODO ####################
    THRESHOLD = 0.3
    N = points_1.shape[0]
    for i in range(N):
        if score_1[i] < THRESHOLD or score_2[i] < THRESHOLD:
            continue
        # Triangulate via DLT
        A = np.zeros((4, 4))
        A[0] = points_1[i, 0] * camera_matrix_1[2] - camera_matrix_1[0]
        A[1] = points_1[i, 1] * camera_matrix_1[2] - camera_matrix_1[1]
        A[2] = points_2[i, 0] * camera_matrix_2[2] - camera_matrix_2[0]
        A[3] = points_2[i, 1] * camera_matrix_2[2] - camera_matrix_2[1]

        _, _, V = np.linalg.svd(A)
        X_homogeneous = V[-1]
        X_euclidean = (
            X_homogeneous[:3] / X_homogeneous[3]
        )  # Normalize to change to euclidian
        output_dict[i] = np.array([X_euclidean[:3]])
    ##############################################

    return output_dict

#################### DO NOT MODIFFY CODES BELOW ####################
data_dir = Path("/content/CV_HW4_tools/test_set") / data_name

# 0.0. load cam data
cam_pth = data_dir / 'floored_scaled_transform.npz'
assert cam_pth.exists(), f"Camera data not found at {str(cam_pth)}"
intr_pth = data_dir / 'undistorted_intrinsics.json'
assert intr_pth.exists(), f"Intrinsic data not found at {str(intr_pth)}"
cam_data = load_cam_data(cam_pth, intr_pth)

# 0.1. load keypoint data
keypoint_dir = Path("/content/CV_HW4_tools/test_set") / data_name / "dwpose" / "kpts"
kp_path_list = sorted(list(keypoint_dir.glob("*.pkl")))
kp_dict = {}

for next_kp_path in kp_path_list:
    cam_idx = int(os.path.basename(next_kp_path)[:-4])
    with open(next_kp_path, "rb") as f:
        k = pickle.load(f)
    body_kp = np.array(k['body_keypoints'][0])
    kp_dict[cam_idx] = body_kp

# 2. two_view_triangulation (Here we selected cam 7 and 27)
cam_mat_7 = cam_data[7]['intrinsics'] @ (np.linalg.inv(cam_data[7]['c2w'])[:3])
cam_mat_27 = cam_data[27]['intrinsics'] @ (np.linalg.inv(cam_data[27]['c2w'])[:3])

kp_7 = kp_dict[7][:, :2]
score_7 = kp_dict[7][:, 2]
kp_27 = kp_dict[27][:, :2]
score_27 = kp_dict[27][:, 2]

two_view_triangulated = two_view_triangulation(cam_mat_7, cam_mat_27, kp_7, kp_27, score_7, score_27)

np.save(data_dir / "two_view_triangulated.npy", two_view_triangulated)
#################### DO NOT MODIFFY CODES BELOW ####################
````

### 2.2. N-view triangulation

````python
import pickle
import os
import torch
import numpy as np
from pathlib import Path

from utils import load_cam_data


def N_view_naive_triangulation(camera_matrices_N, points_N, scores_N):
    """
        input
        camera_matrices_N: N X 3 X 4, N camera matrices of each camera
        points_N: N X M X 2, M joint coordinate in 2D in N views
        scores_N: N X M, M joint score in N views

        output
        output_dict: dictionary of triangulated points
        key: point_idx from 0 to 18, value: triangulated point, np.array with shape 1 X 3
        output_dict = {
            0: [[x1, y1, z1]],
            1: [[x2, y2, z2]],
            ...
            18: [x18, y18, z18]
          }
    """
    output_dict = {}

    #################### TODO ####################
    THRESHOLD = 0.3
    num_views = points_N.shape[0]
    num_joints = points_N.shape[1]

    for joint_idx in range(num_joints):
        # Construct the A matrix for this joint across all views
        A = np.zeros((num_views * 2, 4))
        valid_views = 0

        for view_idx in range(num_views):
            if scores_N[view_idx, joint_idx] < THRESHOLD:
                continue

            u, v = points_N[view_idx, joint_idx]
            P = camera_matrices_N[view_idx]

            # Add two rows to the A matrix for this view
            A[valid_views * 2] = u * P[2] - P[0]
            A[valid_views * 2 + 1] = v * P[2] - P[1]

            valid_views += 1

        # triangulate with atleast 2 valid views
        if valid_views >= 2:
            # Use only the valid rows of the A matrix
            A_valid = A[: valid_views * 2, :]

            # Solve A_valid * X = 0 using SVD
            _, _, V = np.linalg.svd(A_valid)

            # The 3D point is the last column of V
            X_homogeneous = V[-1]
            X_euclidean = (
                X_homogeneous[:3] / X_homogeneous[3]
            )  # Normalize to change to euclidian
            output_dict[joint_idx] = np.array([X_euclidean[:3]])

    ##############################################

    return output_dict

#################### DO NOT MODIFFY CODES BELOW ####################
data_dir = Path("/content/CV_HW4_tools/test_set") / data_name

# 0.0. load cam data
cam_pth = data_dir / 'floored_scaled_transform.npz'
assert cam_pth.exists(), f"Camera data not found at {str(cam_pth)}"
intr_pth = data_dir / 'undistorted_intrinsics.json'
assert intr_pth.exists(), f"Intrinsic data not found at {str(intr_pth)}"
cam_data = load_cam_data(cam_pth, intr_pth)

# 1. load keypoint data
keypoint_dir = Path("/content/CV_HW4_tools/test_set") / data_name / "dwpose" / "kpts"
kp_path_list = sorted(list(keypoint_dir.glob("*.pkl")))
kp_dict = {}

for next_kp_path in kp_path_list:
    cam_idx = int(os.path.basename(next_kp_path)[:-4])
    with open(next_kp_path, "rb") as f:
        k = pickle.load(f)
    body_kp = np.array(k['body_keypoints'][0])
    kp_dict[cam_idx] = body_kp

# 3. N_view_triangulation

cam_mat_total = []
for cam_idx in kp_dict.keys():
    value = cam_data[cam_idx]
    cam_mat_process = value['intrinsics'] @ (np.linalg.inv(value['c2w'])[:3])
    cam_mat_total.append(cam_mat_process)
cam_mat_total = np.array(cam_mat_total)

kp_total = []
score_total = []

for next_kp in kp_dict.values():
    kp_total.append(next_kp[:, :2])
    score_total.append(next_kp[:, 2])

kp_total = np.array(kp_total)
score_total = np.array(score_total)

N_view_triangulated = N_view_naive_triangulation(cam_mat_total, kp_total, score_total)

np.save(data_dir / "N_view_triagulated.npy", N_view_triangulated)
#################### DO NOT MODIFFY CODES BELOW ####################
````

### 3.1. Two-view brute-force triangulation

````python
import pickle
import os
import torch
import numpy as np
from pathlib import Path

from utils import load_cam_data


def calculate_reprojection_error(point_2d_orig, point_3d, camera_matrix):
    point_3d_homogeneous = np.append(point_3d, 1)
    point_2d_homogeneous = camera_matrix @ point_3d_homogeneous
    # Normalize to get Cartesian coordinates
    point_2d_reproj = point_2d_homogeneous[:2] / point_2d_homogeneous[2]
    return np.linalg.norm(point_2d_orig - point_2d_reproj)



def bruteforce_two_view(camera_matrices_2, points_2, scores_2):
    """
        input
        camera_matrices_2: 2 X 3 X 4, camera matrices of each camera
        points_2: 2 X N X M X 2, 2 people's M joint coordinate in 2D in 2 views
        scores_2: 2 X N X M, 2 people's M joint score in N views

        output
        output_dict: key consists of [0, 1], value consists of dictionary of each person's triangulated point like in the format of 2-1 and 2-2
        output_dict = {
          0: {
            0: [[x1, y1, z1]],
            1: [[x2, y2, z2]],
            ...
            18: [x18, y18, z18]
          }
          1: {
            ...
          }
        }

    """
    output_dict = {}


    #################### TODO ####################
    THRESHOLD = 0.3
    num_joints = points_2.shape[2]

    cam_mat_1 = camera_matrices_2[0]
    cam_mat_2 = camera_matrices_2[1]

    # Points and scores for view 2
    points_v2_p1 = points_2[0, 1]
    points_v2_p2 = points_2[1, 1]
    scores_v2_p1 = scores_2[0, 1]
    scores_v2_p2 = scores_2[1, 1]

    # Define the two possible pairings for view 2
    pairings = [
        # Pairing 1: person 1 in view 1 corresponds to person 1 in view 2
        #            person 2 in view 1 corresponds to person 2 in view 2
        ([points_v2_p1, points_v2_p2], [scores_v2_p1, scores_v2_p2]),
        # Pairing 2: person 1 in view 1 corresponds to person 2 in view 2
        #            person 2 in view 1 corresponds to person 1 in view 2
        ([points_v2_p2, points_v2_p1], [scores_v2_p2, scores_v2_p1]),
    ]

    min_avg_reproj_error = float("inf")

    for (points_v2_paired, scores_v2_paired) in pairings:
        current_triangulated_points = {}
        total_reproj_error = 0
        valid_joint_count = 0

        for person_idx in range(2):  # Iterate through the two people in view 1
            person_output_dict = {}
            points_v1 = points_2[person_idx, 0]
            scores_v1 = scores_2[person_idx, 0]

            points_v2 = points_v2_paired[person_idx]
            scores_v2 = scores_v2_paired[person_idx]

            for joint_idx in range(num_joints):
                if scores_v1[joint_idx] < THRESHOLD or scores_v2[joint_idx] < THRESHOLD:
                    continue

                p1 = points_v1[joint_idx]
                p2 = points_v2[joint_idx]

                # Triangulate the point
                # Construct the A matrix for the system of linear equations AX = 0
                A = np.zeros((4, 4))

                A[0] = p1[0] * cam_mat_1[2] - cam_mat_1[0]
                A[1] = p1[1] * cam_mat_1[2] - cam_mat_1[1]
                A[2] = p2[0] * cam_mat_2[2] - cam_mat_2[0]
                A[3] = p2[1] * cam_mat_2[2] - cam_mat_2[1]

                _, _, V = np.linalg.svd(A)
                X_homogeneous = V[-1]
                triangulated_point = X_homogeneous[:3] / X_homogeneous[3]
                person_output_dict[joint_idx] = np.array([triangulated_point[:3]])

                # Calculate and accumulate reprojection error
                error_v1 = calculate_reprojection_error(
                    p1, triangulated_point, cam_mat_1
                )
                error_v2 = calculate_reprojection_error(
                    p2, triangulated_point, cam_mat_2
                )
                total_reproj_error += (error_v1 + error_v2) / 2
                valid_joint_count += 1

            current_triangulated_points[person_idx] = person_output_dict

        # Calculate average reprojection error for this pairing
        avg_reproj_error = total_reproj_error / (
            valid_joint_count if valid_joint_count > 0 else 1
        )
        if avg_reproj_error < min_avg_reproj_error:
            min_avg_reproj_error = avg_reproj_error
            output_dict = current_triangulated_points
    ##############################################

    return output_dict

#################### DO NOT MODIFFY CODES BELOW ####################
data_dir = Path("/content/CV_HW4_tools/test_set") / data_name

# 0.0. load cam data
cam_pth = data_dir / 'floored_scaled_transform.npz'
assert cam_pth.exists(), f"Camera data not found at {str(cam_pth)}"
intr_pth = data_dir / 'undistorted_intrinsics.json'
assert intr_pth.exists(), f"Intrinsic data not found at {str(intr_pth)}"
cam_data = load_cam_data(cam_pth, intr_pth)

# 1. load hpe output
keypoint_dir = Path("/content/CV_HW4_tools/test_set") / data_name / "dwpose" / "kpts"

kp_path_list = sorted(keypoint_dir.glob("*.pkl"))

kp_dict = {}

for next_kp_path in kp_path_list:
    cam_idx = int(os.path.basename(next_kp_path)[:-4])
    with open(next_kp_path, "rb") as f:
        k = pickle.load(f)
    if len(k['body_keypoints']) < 2:
        continue
    body_kp_1 = np.array(k['body_keypoints'][0])
    body_kp_2 = np.array(k['body_keypoints'][1])
    kp_dict[cam_idx] = [body_kp_1, body_kp_2]

# 2. two_view_triangulation

cam_mat_9 = cam_data[9]['intrinsics'] @ (np.linalg.inv(cam_data[9]['c2w'])[:3])
cam_mat_24 = cam_data[24]['intrinsics'] @ (np.linalg.inv(cam_data[24]['c2w'])[:3])

cam_mat_two_view = np.stack([cam_mat_9, cam_mat_24])

kp_9_1 = kp_dict[9][0][:, :2]
kp_9_2 = kp_dict[9][1][:, :2]
score_9_1 = kp_dict[9][0][:, 2]
score_9_2 = kp_dict[9][1][:, 2]
kp_24_1 = kp_dict[24][0][:, :2]
kp_24_2 = kp_dict[24][1][:, :2]
score_24_1 = kp_dict[24][0][:, 2]
score_24_2 = kp_dict[24][1][:, 2]

kp_two_p_two_view = np.zeros((2, 2, 19, 2))
kp_two_p_two_view[0, 0] = kp_9_1
kp_two_p_two_view[0, 1] = kp_24_1
kp_two_p_two_view[1, 0] = kp_9_2
kp_two_p_two_view[1, 1] = kp_24_2

score_two_view = np.zeros((2, 2, 19))
score_two_view[0, 0] = score_9_1
score_two_view[0, 1] = score_24_1
score_two_view[1, 0] = score_9_2
score_two_view[1, 1] = score_24_2

two_view_brute = bruteforce_two_view(cam_mat_two_view, kp_two_p_two_view, score_two_view)

np.save(data_dir / "two_view_brute.npy", two_view_brute)
#################### DO NOT MODIFFY CODES BELOW ####################
````

## 3.2. N-view brute triangulation

````python
def triangulate_point_from_views(camera_matrices, points_2d):
    if len(camera_matrices) < 2:
        return None

    num_views = len(camera_matrices)
    A = np.zeros((num_views * 2, 4))

    for i in range(num_views):
        P = camera_matrices[i]
        u, v = points_2d[i]

        A[i * 2] = u * P[2] - P[0]
        A[i * 2 + 1] = v * P[2] - P[1]

    _, _, V = np.linalg.svd(A)
    X_homogeneous = V[-1]
    X_euclidean = X_homogeneous[:3] / X_homogeneous[3]
    return X_euclidean
````

````python
import pickle
import os
import torch
import numpy as np
from pathlib import Path
from tqdm import tqdm
import random

from utils import load_cam_data

def N_view_greedy(camera_matrices_N, points_N, scores_N):
    """
        input
        camera_matrices_N: N X 3 X 4, camera matrices of each camera
        points_N: 2 X N X M X 2, 2 people's M joint coordinate in 2D in N views
        scores_N: 2 X N X M, 2 people's M joint score in N views

        output
        output_dict: key consists of [0, 1], value consists of dictionary of each person's triangulated point like in the format of 2-1 and 2-2
        output_dict = {
          0: {
            0: [[x1, y1, z1]],
            1: [[x2, y2, z2]],
            ...
            18: [x18, y18, z18]
          }
          1: {
            ...
          }
        }
    """
    output_dict = {}
    #################### TODO ####################
    MAX_ITERATIONS = 10000
    THRESHOLD = 0.3
    num_views = points_N.shape[1]
    num_joints = points_N.shape[2]
    min_avg_reproj_error = float("inf")
    best_pairing = None

    available_cam_indices = list(range(num_views))

    for _ in tqdm(range(MAX_ITERATIONS), desc="Sampling Pairings"):
        sampled_cam_indices = random.sample(available_cam_indices, 2)
        cam_idx1, cam_idx2 = sampled_cam_indices

        # Assume the keypoints from these two views are the base for a potential pairing
        points_v2_p1 = points_N[0, cam_idx2]
        points_v2_p2 = points_N[1, cam_idx2]
        scores_v2_p1 = scores_N[0, cam_idx2]
        scores_v2_p2 = scores_N[1, cam_idx2]

        # Define the two possible pairings for the sampled views
        pairings = [
            (
                [points_v2_p1, points_v2_p2],
                [scores_v2_p1, scores_v2_p2],
                [cam_idx1, cam_idx2],
            ),  # Pairing 1: person 1 in view 1 corresponds to person 1 in view 2
            (
                [points_v2_p2, points_v2_p1],
                [scores_v2_p2, scores_v2_p1],
                [cam_idx1, cam_idx2],
            ),  # Pairing 2: person 1 in view 1 corresponds to person 2 in view 2
        ]

        for pairing_idx, (
            points_v2_paired,
            scores_v2_paired,
            used_cam_indices,
        ) in enumerate(pairings):
            current_triangulated_points = {}
            total_reproj_error = 0
            valid_joint_count = 0

            for person_idx in range(
                2
            ):  # Iterate through the two people based on the first sampled view's detection
                person_output_dict = {}

                for joint_idx in range(num_joints):
                    # Collect valid 2D points and camera matrices for this joint across all views
                    valid_points_2d = []
                    valid_camera_matrices = []
                    valid_views_for_joint = []

                    # Check the two initially sampled views based on the current pairing
                    for view_pair_idx in range(2):
                        cam_idx = used_cam_indices[view_pair_idx]
                        if view_pair_idx == 0:
                            points_v = points_N[person_idx, cam_idx]
                            scores_v = scores_N[person_idx, cam_idx]
                        else:  # For the second view in the pairing, use the potentially swapped detection
                            points_v = points_v2_paired[person_idx]
                            scores_v = scores_v2_paired[person_idx]

                        if scores_v[joint_idx] > THRESHOLD:
                            valid_points_2d.append(points_v[joint_idx])
                            valid_camera_matrices.append(camera_matrices_N[cam_idx])
                            valid_views_for_joint.append(cam_idx)

                    # Triangulate the point if there are at least two valid views
                    triangulated_point = None
                    if len(valid_points_2d) >= 2:
                        triangulated_point = triangulate_point_from_views(
                            valid_camera_matrices, valid_points_2d
                        )

                    if triangulated_point is not None:
                        person_output_dict[joint_idx] = np.array(
                            [triangulated_point[:3]]
                        )

                        # Calculate and accumulate reprojection error for this joint across all views used for triangulation
                        joint_reproj_error = 0
                        for i, cam_idx in enumerate(valid_views_for_joint):
                            # Need to get the original 2D point for this person and camera
                            original_point_2d = None
                            if cam_idx == used_cam_indices[0]:
                                original_point_2d = points_N[
                                    person_idx, cam_idx, joint_idx
                                ]
                            elif cam_idx == used_cam_indices[1]:
                                # If it's the second view in the pairing, we need to consider which detection it was matched with
                                if (
                                    pairing_idx == 0
                                ):  # Person 1 in view 1 matched with Person 1 in view 2, Person 2 with Person 2
                                    original_point_2d = points_N[
                                        person_idx, cam_idx, joint_idx
                                    ]
                                else:  # Person 1 in view 1 matched with Person 2 in view 2, Person 2 with Person 1
                                    original_point_2d = points_N[
                                        1 - person_idx, cam_idx, joint_idx
                                    ]

                            if original_point_2d is not None:
                                error = calculate_reprojection_error(
                                    original_point_2d,
                                    triangulated_point,
                                    camera_matrices_N[cam_idx],
                                )
                                joint_reproj_error += error

                        if len(valid_views_for_joint) > 0:
                            total_reproj_error += joint_reproj_error / len(
                                valid_views_for_joint
                            )
                            valid_joint_count += 1

                current_triangulated_points[person_idx] = person_output_dict

            # Calculate average reprojection error for this pairing
            avg_reproj_error = total_reproj_error / (
                valid_joint_count if valid_joint_count > 0 else 1
            )
            if avg_reproj_error < min_avg_reproj_error:
                min_avg_reproj_error = avg_reproj_error
                best_pairing = current_triangulated_points

    output_dict = best_pairing
    ##############################################

    return output_dict


#################### DO NOT MODIFFY CODES BELOW ####################
data_dir = Path("/content/CV_HW4_tools/test_set") / data_name

# 0.0. load cam data
cam_pth = data_dir / 'floored_scaled_transform.npz'
assert cam_pth.exists(), f"Camera data not found at {str(cam_pth)}"
intr_pth = data_dir / 'undistorted_intrinsics.json'
assert intr_pth.exists(), f"Intrinsic data not found at {str(intr_pth)}"
cam_data = load_cam_data(cam_pth, intr_pth)

# 1. load hpe output
keypoint_dir = Path("/content/CV_HW4_tools/test_set") / data_name / "dwpose" / "kpts"

kp_path_list = sorted(keypoint_dir.glob("*.pkl"))

kp_dict = {}

for next_kp_path in kp_path_list:
    cam_idx = int(os.path.basename(next_kp_path)[:-4])
    with open(next_kp_path, "rb") as f:
        k = pickle.load(f)
    if len(k['body_keypoints']) < 2:
        continue
    body_kp_1 = np.array(k['body_keypoints'][0])
    body_kp_2 = np.array(k['body_keypoints'][1])
    kp_dict[cam_idx] = [body_kp_1, body_kp_2]

cam_mat_total = []
for cam_idx in kp_dict.keys():
    value = cam_data[cam_idx]
    cam_mat_process = value['intrinsics'] @ (np.linalg.inv(value['c2w'])[:3])
    cam_mat_total.append(cam_mat_process)
cam_mat_total = np.array(cam_mat_total)

kp_total = np.zeros((2, len(cam_mat_total), 19, 2))
score_total = np.zeros((2, len(cam_mat_total), 19))

for view_idx, next_kp in enumerate(kp_dict.values()):
    kp_1 = next_kp[0][:, :2]
    kp_2 = next_kp[1][:, :2]
    score_1 = next_kp[0][:, 2]
    score_2 = next_kp[1][:, 2]
    kp_total[0, view_idx] = kp_1
    kp_total[1, view_idx] = kp_2
    score_total[0, view_idx] = score_1
    score_total[1, view_idx] = score_2

N_view_greedy_output = N_view_greedy(cam_mat_total, kp_total, score_total)

np.save(data_dir / "N_view_brute.npy", N_view_greedy_output)
#################### DO NOT MODIFFY CODES BELOW ####################
````

### 4.1. Node(Joint) Proposals


Note that the heatmap resolution is 1/4 times of original image resolution on both X and Y axis. We provide you NMS function, but it's totally free to implement your own function. Note that the output is numpy array format here.

````python
from pathlib import Path


import torch
import torch.nn.functional as F
import numpy as np

from utils import load_cam_data, N_JNTS

def compute_3d_node_proposals(n_jnts, heatmaps, extrinsics, intrinsics, voxel_grid, transform_matrix, threshold=0.09):
    """
    input
      heatmaps: Dict[cam_id, Tensor(N,J,h,w)]
            - N: # of detected person
            - J: # of joints (19, ==N_JNTS)
            - h: heatmap height (about 375) (1/4 size of original image height)
            - w: heatmap width (about 500) (1/4 size of original image width)
      extrinsics: Dict[cam_id, 4x4 tensor (cam to world)]
      intrinsics: Dict[cam_id, 3x3 tensor]
            - Note that this intrinsic is for image sized about (1500, 2000) (slight different per view). you should handle resolution difference properly
      voxel_grid: Tensor(Gx, Gy, Gz, 3)
            - each voxel holds 3D world coordinate.
      threshold: float, score threshold for NMS

    output
      node_proposals: Dict[joint_idx, np.array(N_j, 3)] Dictionary of Numpy array
        - joint_idx(int): (0~18)
        - N_j: # of proposed nodes corresponding to j-th joint.

    """
    node_proposals = dict()

    #################### TODO ####################
    # compute a merged score map by taking the maximum across all person detections at each pixel
    # h^c_j(z)
    reduced_heatmaps = {}
    for cam_id in heatmaps.keys():
        reduced_heatmaps[cam_id] = heatmaps[cam_id].max(dim=0)[
            0
        ]

    # Get grid dimensions
    Gx, Gy, Gz = voxel_grid.shape[:3]
    N_points = Gx * Gy * Gz
    device = voxel_grid.device

    # Initialize accumulation tensors for all joints
    total_scores = torch.zeros((n_jnts, N_points), device=device)
    counts = torch.zeros(N_points, device=device)
    # Transform voxel grid points to camera-aligned world coordinates
    voxels = voxel_grid.view(-1, 3)
    voxels_hom = torch.cat(
        [voxels, torch.ones(N_points, 1, device=device)], dim=-1
    )
    for cam_id in heatmaps.keys():
        extrinsic = extrinsics[cam_id].to(device)
        intrinsic = intrinsics[cam_id].to(device)

        # Compute world to camera transformation
        world_to_cam = torch.inverse(extrinsic)

        # Transform to camera coordinates
        points_cam_hom = torch.mm(world_to_cam, voxels_hom.T).T

        x, y, z = points_cam_hom[:, 0], points_cam_hom[:, 1], points_cam_hom[:, 2]

        valid_depth = z > 0
        if not valid_depth.any():
            continue

        # Project to image plane
        x_proj = x / z
        y_proj = y / z
        u = intrinsic[0, 0] * x_proj + intrinsic[0, 2]
        v = intrinsic[1, 1] * y_proj + intrinsic[1, 2]

        # Scale to heatmap resolution
        u_heat = u / 4.0
        v_heat = v / 4.0

        # Get heatmap dimensions for this camera
        _, h_heat, w_heat = reduced_heatmaps[cam_id].shape

        # Check bounds
        valid_bounds = (
            (u_heat >= 0)
            & (u_heat <= w_heat - 1)
            & (v_heat >= 0)
            & (v_heat <= h_heat - 1)
        )
        valid = valid_depth & valid_bounds

        # Create normalized grid coordinates for grid_sample
        grid_x = (u_heat / (w_heat - 1)) * 2 - 1
        grid_y = (v_heat / (h_heat - 1)) * 2 - 1
        grid_pts = (
            torch.stack((grid_x, grid_y), dim=1).view(1, 1, -1, 2).to(device)
        )

        # Sample the heatmap for all joints at once
        heatmap_cam_batch = (
            reduced_heatmaps[cam_id].unsqueeze(0).to(device)
        )
        sampled = F.grid_sample(
            heatmap_cam_batch,
            grid_pts,
            mode="bilinear",
            align_corners=True,
            padding_mode="zeros",
        )
        sampled = sampled.squeeze(0).squeeze(1)

        # Update total_scores and counts for valid points
        total_scores[:, valid] += sampled[:, valid]
        counts[valid] += 1

    # Normalize by cardinality of the set
    score_maps = total_scores / counts.clamp(min=1e-10)  # Avoid division by zero
    score_maps = score_maps.view(n_jnts, Gx, Gy, Gz)

    # For each joint, reshape to 3D grid and apply NMS
    node_proposals = {}
    for j in range(n_jnts):
        # Apply NMS
        voxel_idx, peak_scores = nms3d(score_maps[j], tau=threshold, radius=5)

        if voxel_idx.numel() == 0:
            node_proposals[j] = np.zeros((0, 3), dtype=np.float32)
        else:
            proposals_3d = voxel_grid[voxel_idx[:, 0], voxel_idx[:, 1], voxel_idx[:, 2]]
            node_proposals[j] = proposals_3d.cpu().numpy()
    ##############################################

    return node_proposals  # List of tensors [num_nodes x 3] per joint



def nms3d(score_map, tau=0.05, radius=5):
    """
    Non-Maxima Suppression on a 3-D voxel score-map.

    Args:
        score_map (torch.Tensor): (Gx, Gy, Gz) tensor with 3-D node scores.
        tau (float)              : keep only peaks whose score > tau.
        radius (int)             : NMS window radius ⇒ window = (2r+1)^3.

    Returns:
        voxel_idx (N×3 long) — indices in the voxel grid (z,y,x).
        peak_scores (N,)     — their scores.
    """

    # (1,1,D,H,W) so we can use max_pool3d
    s = score_map.unsqueeze(0).unsqueeze(0)

    # local max within a cubic window
    pooled = F.max_pool3d(
        s, kernel_size=2 * radius + 1, stride=1, padding=radius
    )

    keep = (s == pooled) & (s > tau)                 # boolean mask for peaks
    voxel_idx = keep.nonzero(as_tuple=False)[:, -3:]  # drop batch/channel dims

    if voxel_idx.numel() == 0:
        return (
            voxel_idx,
            torch.empty((0,), device=score_map.device),
        )

    peak_scores = score_map[voxel_idx[:, 0],
                            voxel_idx[:, 1],
                            voxel_idx[:, 2]]

    return voxel_idx, peak_scores



def make_voxel_grid(c2ws, grid_size=0.02, grid_shape=(1.0, 1.0, 1.0), device='cuda'):
    """
    Create a 3-D voxel grid in world coordinates.

    1. From camera locations, define large 3D box including all cameras with enough margins.
    2. Then divide the 3D box into small voxel cubes whose grid_size is as given. (0.02m == 2cm)

    Note that z=0 is ground and -z is up direction in calibrated cameras here! We HIGHLY recommend
    you to implement the box include all  -0.1 < z < 2.0  (as person height < 2.0m in general).


    Args:
        c2ws (Dict[int, torch.Tensor]): {cam_id: 4×4 cam-to-world matrix}.
        grid_size (float): edge length of each voxel (meters).
        grid_shape (Tuple[float,float,float]): physical size of the grid (X,Y,Z, meters).
        device (str): 'cuda' or 'cpu'.

    Returns:
        voxel_grid (torch.Tensor): (Nx,Ny,Nz,3) world-space coordinates.
        transform_matrix (torch.Tensor): 4×4 matrix that maps voxel indices → world coords.
    """


    #################### TODO ####################
    cam_positions = torch.stack([c2w[:3, 3] for c2w in c2ws.values()], dim=0).to(device)

    # Find center of camera positions
    center = torch.mean(cam_positions, dim=0)

    # Define grid dimensions using grid_shape, centered at camera positions
    half_size = torch.tensor(grid_shape, device=device) / 2.0
    min_coords = center - half_size
    max_coords = center + half_size

    # Ensure z bounds cover the specified range (-0.1 to 2.0)
    min_coords[2] = min(min_coords[2].item(), -0.1)
    max_coords[2] = max(max_coords[2].item(), 2.0)

    # Calculate the number of voxels in each dimension
    Nx = int((max_coords[0] - min_coords[0]) / grid_size) + 1
    Ny = int((max_coords[1] - min_coords[1]) / grid_size) + 1
    Nz = int((max_coords[2] - min_coords[2]) / grid_size) + 1

    # Create transformation matrix from voxel indices to world coordinates
    transform_matrix = torch.eye(4, device=device)
    transform_matrix[:3, 3] = min_coords
    transform_matrix[:3, :3] = torch.diag(
        torch.tensor([grid_size, grid_size, grid_size], device=device)
    )

    # Create grid of indices
    x = torch.arange(Nx, device=device)
    y = torch.arange(Ny, device=device)
    z = torch.arange(Nz, device=device)

    # Create 3D grid of indices
    grid_x, grid_y, grid_z = torch.meshgrid(x, y, z, indexing="ij")

    # Create homogeneous coordinates for each voxel
    indices = torch.stack(
        [grid_x, grid_y, grid_z, torch.ones_like(grid_x)], dim=-1
    ).float()

    # Reshape for matrix multiplication
    indices_flat = indices.reshape(-1, 4)

    # Apply transformation to get world coordinates
    world_coords_flat = torch.matmul(indices_flat, transform_matrix.T)

    # Reshape back to grid structure (Nx, Ny, Nz, 3)
    voxel_grid = world_coords_flat[:, :3].reshape(Nx, Ny, Nz, 3)
    ##############################################

    return voxel_grid, transform_matrix




#################### DO NOT MODIFFY CODES BELOW ####################
data_dir = Path("/content/CV_HW4_tools/test_set") / data_name

# 0.0. load heatmaps
heatmap_dir = data_dir / "dwpose" / 'heatmaps'
assert heatmap_dir.exists(), f"Heatmap data not found at {str(heatmap_dir)}"
heatmap_fdict = dict()
for heatmap_fname in heatmap_dir.glob("*.npy"):
    cam_id = int(heatmap_fname.stem.split('.')[0])
    heatmap_fdict[cam_id] = heatmap_fname

# 0.1. get cameras
cam_pth = data_dir / 'floored_scaled_transform.npz'
assert cam_pth.exists(), f"Camera data not found at {str(cam_pth)}"
intr_pth = data_dir / 'undistorted_intrinsics.json'
assert intr_pth.exists(), f"Intrinsic data not found at {str(intr_pth)}"
cam_data = load_cam_data(cam_pth, intr_pth)
extrinsics = {cam_id: torch.from_numpy(cam_data[cam_id]['c2w']) for cam_id in sorted(list(cam_data.keys()))}
intrinsics = {cam_id: torch.from_numpy(cam_data[cam_id]['intrinsics']) for cam_id in sorted(list(cam_data.keys()))}

# 1. get voxel grid of scene
voxel_grid, transform_matrix = make_voxel_grid(c2ws=extrinsics)

# 2. compute 3D node proposals
body_heatmap_dict = {cam_id: torch.from_numpy(np.load(heatmap_fdict[cam_id])[:N_JNTS]) for cam_id in heatmap_fdict}   # Assume 24 joints

node_proposals = compute_3d_node_proposals(
    n_jnts=N_JNTS,  # Number of joints in the body
    heatmaps=body_heatmap_dict,
    extrinsics=extrinsics,
    intrinsics=intrinsics,
    voxel_grid=voxel_grid,
    transform_matrix=transform_matrix,
)
print(f"Node proposals computed for {len(node_proposals)} joints.")

save_dir = data_dir / "panoptic"
save_dir.mkdir(parents=True, exist_ok=True)
np.save(save_dir / "node_proposals.npy", node_proposals)

#################### DO NOT MODIFFY CODES ABOVE ####################
````

### 4.2. Part(Bone) Proposals

Note that the heatmap resolution is 1/4 times of original image resolution on both X and Y axis. Also be aware that output part score list has numpy array as element.

````python
import pickle
from pathlib import Path

import torch
import torch.nn.functional as F
import numpy as np


from utils import load_cam_data, get_bones, create_bones


def generate_part_proposals(node_proposals, joint_pairs, heatmaps, extrinsics, intrinsics, threshold=0.05):
    """
    input
      node_proposals: Dict[joint_idx, Tensor(N_j, 3)]
        - joint_idx(int): (0~18)
        - N_j: # of proposed nodes corresponding to j-th joint.
      joint_pairs: [(u,v),...] bones defined by joint indices(int) of each bone.
      heatmaps: Dict[cam_id, Tensor(N,J,h,w)]
            - N: # of detected person
            - J: # of joints (19, ==N_JNTS)
            - h: heatmap height (=375) (1/4 size of original image height)
            - w: heatmap width (=502) (1/4 size of original image width)
      extrinsics: Dict[cam_id, 4x4 tensor (cam to world)]
      intrinsics: Dict[cam_id, 3x3 tensor]
            - Note that this intrinsic is for image sized (1500, 2008). you should mathc them properly
    output
      part_scores: List[np.array(N_u, N_v)] List of score maps
        - N_u: # of node proposals of u-th joint
        - N_v: # of node proposals of v-th joint

    """

    part_scores = []

    for u, v in joint_pairs:
        #################### TODO ####################
        # Get node proposals for this joint pair
        nodes_u = node_proposals[u]  # N_u x 3
        nodes_v = node_proposals[v]  # N_v x 3

        N_u = nodes_u.shape[0]
        N_v = nodes_v.shape[0]

        # Initialize score matrix and view counter
        score_matrix = np.zeros((N_u, N_v))
        view_counts = np.zeros((N_u, N_v))

        # For each camera view
        for cam_id, hmap in heatmaps.items():
            # Get camera parameters
            extrinsic = extrinsics[cam_id]  # 4x4 cam to world
            intrinsic = intrinsics[cam_id]  # 3x3

            # Convert cam-to-world to world-to-cam for projection
            world_to_cam = torch.inverse(extrinsic)

            # Get heatmap dimensions
            h, w = hmap.shape[2], hmap.shape[3]

            # Calculate scale factors from original image to heatmap
            scale_x = 2008 / w  # original width / heatmap width
            scale_y = 1500 / h  # original height / heatmap height

            # For each pair of node proposals
            for ui in range(N_u):
                for vi in range(N_v):
                    node_u = nodes_u[ui]
                    node_v = nodes_v[vi]

                    # Convert to homogeneous coordinates
                    node_u_homo = torch.cat([node_u, torch.tensor([1.0])], dim=0)
                    node_v_homo = torch.cat([node_v, torch.tensor([1.0])], dim=0)

                    # Transform to camera space
                    cam_u_homo = world_to_cam @ node_u_homo
                    cam_v_homo = world_to_cam @ node_v_homo
                    cam_u = cam_u_homo[:3]
                    cam_v = cam_v_homo[:3]

                    # Skip if either point is behind the camera
                    if cam_u[2] <= 0 or cam_v[2] <= 0:
                        continue

                    # Project to image plane
                    img_u_homo = intrinsic @ (cam_u / cam_u[2])
                    img_v_homo = intrinsic @ (cam_v / cam_v[2])
                    img_u = img_u_homo[:2]
                    img_v = img_v_homo[:2]

                    # Scale to heatmap resolution
                    hmap_u = torch.tensor([img_u[0] / scale_x, img_u[1] / scale_y])
                    hmap_v = torch.tensor([img_v[0] / scale_x, img_v[1] / scale_y])
                    hmap_u_int = torch.round(hmap_u).long()
                    hmap_v_int = torch.round(hmap_v).long()

                    # Check if points are within heatmap bounds
                    if (
                        0 <= hmap_u_int[0] < w
                        and 0 <= hmap_u_int[1] < h
                        and 0 <= hmap_v_int[0] < w
                        and 0 <= hmap_v_int[1] < h
                    ):
                        view_counts[ui, vi] += 1

                        # Find maximum score across all person detections
                        max_score = 0
                        num_people = hmap.shape[0]

                        for i in range(num_people):
                            # Get heatmap values at projected locations
                            score_u = hmap[i, u, hmap_u_int[1], hmap_u_int[0]].item()
                            score_v = hmap[i, v, hmap_v_int[1], hmap_v_int[0]].item()

                            # Apply delta function - both nodes must be detected as part of same person
                            if score_u > threshold and score_v > threshold:
                                # Calculate part connectivity score (average of node scores)
                                w_c_iuv = (score_u + score_v) / 2

                                # Take maximum across all people
                                max_score = max(max_score, w_c_iuv)

                        # Add the score from this view
                        score_matrix[ui, vi] += max_score

        # Average across all views where the part is visible
        valid_pairs = view_counts > 0
        if valid_pairs.any():
            score_matrix[valid_pairs] /= view_counts[valid_pairs]

        part_scores.append(score_matrix)
        ##############################################

    return part_scores  # List of numpy arrays for each joint pair





def project_points(points, K, E, hm_shape):
    """
    input
      points: Tensor(N, 3) 3D world coordinates
      K: Tensor(3,3) Intrinsic
      E: Tensor(4,4) Extrinsic (world to cam)
      hm_shape: Tensor(2): heatmap size (h, w)

    output:
      norm_2d: Tensor(N, 2) normalized 2D coordiante of projected keypoint in [-1, 1].

    As we provide the heatmap as 1/4 resized heatmap, we need to this customized
    function to get accurate normalized 2D coordinate of projected keypoints. It's
    fine to use your own function.
    """
    points_homo = torch.cat([points, torch.ones(*points.shape[:-1], 1, device=points.device)], dim=-1)
    points_cam = (E @ points_homo.view(-1,4).T).T[:, :3]
    proj_2d = (K @ points_cam.T).T
    proj_2d = proj_2d[:, :2] / proj_2d[:, 2:3]

    H, W = hm_shape
    H = H * 4
    W = W * 4

    norm_2d = proj_2d.clone()
    norm_2d[:, 0] = (norm_2d[:, 0] / (W - 1)) * 2 - 1
    norm_2d[:, 1] = (norm_2d[:, 1] / (H - 1)) * 2 - 1

    return norm_2d





#################### DO NOT MODIFFY CODES BELOW ####################
data_dir = Path("/content/CV_HW4_tools/test_set") / data_name

# 0.0. load dwpose outputs
heatmap_dir = data_dir / "dwpose" / 'heatmaps'
assert heatmap_dir.exists(), f"Heatmap data not found at {str(heatmap_dir)}"
heatmap_fdict = dict()
for heatmap_fname in heatmap_dir.glob("*.npy"):
    cam_id = int(heatmap_fname.stem.split('.')[0])
    heatmap_fdict[cam_id] = heatmap_fname

# 0.1. get cameras
cam_pth = data_dir / 'floored_scaled_transform.npz'
assert cam_pth.exists(), f"Camera data not found at {str(cam_pth)}"
intr_pth = data_dir / 'undistorted_intrinsics.json'
assert intr_pth.exists(), f"Intrinsic data not found at {str(intr_pth)}"
cam_data = load_cam_data(cam_pth, intr_pth)

extrinsics = {cam_id: torch.from_numpy(cam_data[cam_id]['c2w']) for cam_id in sorted(list(cam_data.keys()))}
intrinsics = {cam_id: torch.from_numpy(cam_data[cam_id]['intrinsics']) for cam_id in sorted(list(cam_data.keys()))}


# 0.2. load node proposals
node_proposals = np.load(data_dir / "panoptic" / "node_proposals.npy", allow_pickle=True).item()
node_proposals = {int(k): torch.tensor(v, dtype=torch.float32) for k, v in node_proposals.items()}

# 0.3. get bone connections
bones = get_bones()

# 1. Get Part Proposals
part_scores = generate_part_proposals(
    node_proposals=node_proposals,
    joint_pairs=bones,
    heatmaps={cam_id: torch.from_numpy(np.load(heatmap_fdict[cam_id])) for cam_id in heatmap_fdict},
    extrinsics=extrinsics,
    intrinsics=intrinsics,
    threshold=0.05
)

# save part proposals
save_dir = data_dir / "panoptic"
with open(save_dir / 'part_scores.pkl', 'wb') as f:
    pickle.dump(part_scores, f)

# save bones for viz
viz_parts = create_bones(
    part_scores=part_scores,
    joint_pairs=bones,
    node_proposals=node_proposals,
    viz_thrs=0.01
)
np.save(save_dir / "viz_part_proposals.npy", viz_parts)
#################### DO NOT MODIFFY CODES ABOVE ####################
````
